```python
import pandas as pd
url='https://raw.githubusercontent.com/LuciaMaRu/df_sucio/refs/heads/main/df_sucio.csv'
df=pd.read_csv(url)
df
```

```python
# resumen estadÃ­stico de los datos.
df.describe()
```

```python
#Valores faltantes
df.isnull().sum()
```

```python
#Filas duplicadas
df.duplicated().sum()
```

```python
#Tipos de datos
df.info()
```

```python
#LIMPIEZA DE DATOS
```

```python
#Eliminar filas duplicadas
df1=df.drop_duplicates()
df1
```

```python
#Eliminar invalid
lista_col=df1.columns
df2=df1
for i in lista_col:
    df2=df2[df2[i] != 'invalid']
df2
```

```python
#Valores faltantes
df3=df2.dropna(subset=['Date'])
df3
```

```python
#Cambiar formatos
df3.columns
```

```python
df3['Date'] = pd.to_datetime(df3['Date'])
lista2_col=['High', 'Low', 'Close', 'Adj Close']
for i in lista2_col:
    df3[i]=df3[i].astype(float)
df3.info()
```

```python
#Valores faltantes
df4=df3
lista1=['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']
for i in lista1:
    df3[i].fillna(df3[i].mean(), inplace=True)
df4
```

```python
df4.isnull().sum()
```

```python
df4.describe()
```

```python
df4
```

```python
#Ordenar los datos por fecha
df5=df4.sort_values(by=['Date'])
df5
```

```python
#Reindexar
df6 = df5.reset_index(drop=True)
df6
```

```python
df6.isnull().sum()
```

```python
df6.info()
```

```python
df6.duplicated().sum()
```

```python
#Guardar CSV
df6.to_csv("Base_limpia.csv", index=True)
```
